**Scaling**

- Scaling is the one of most importent step before model development

- Scaling means makes all the columns under one scale

- Scaling used to make all the columns or features comparable

- Some ML models works on Distance methods 

     - Example age min : 0  max:100

     - Income might be lakhs crores so much bigger values

     - If we dont make age and income under one scale ML model treats Income is the important variable

     - When value are huge maths makes more complex so it is better to do lower down the values

     - All the features under one scale so easy to compare

        - Dollors and Rupees we can not compare because two are different scales

**Standard Scalar**

   - z scale makes mean=0 and standard deviation always=1

    <img src="https://www.inchcalculator.com/wp-content/uploads/2022/11/z-score-formula.png" id="img" style="width: 718px; height: 479px;">

```python
x=visa_df['prevailing_wage']
x_mean=visa_df['prevailing_wage'].mean()
x_std=visa_df['prevailing_wage'].std()
wage_z=(x-x_mean)/x_std
```

```python
wage_z.mean(),wage_z.std()
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1).hist(x)
plt.subplot(1,2,2).hist(wage_z)
plt.show()
```
