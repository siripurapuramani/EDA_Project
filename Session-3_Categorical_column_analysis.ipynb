**Bar Chart**

```python
import matplotlib.pyplot as plt
import pandas as pd
file_path=pd.read_csv(r"C:\Users\ACER\Downloads\Visadataset.csv")
plt.figure(figsize=(9,5))
plt.xlabel('Contents')
plt.ylabel('Number of Applicants')
colors = ['red', 'green', 'blue','orange','yellow','lightgreen']
keys=file_path['continent'].value_counts().keys()
val=file_path['continent'].value_counts().values
bars = plt.bar(keys, val, color=colors)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(height), 
             ha='center', va='bottom', fontsize=10)
plt.title('Continent Bar Chart')
plt.savefig('content_bar_chart.jpg')
plt.show()
```
